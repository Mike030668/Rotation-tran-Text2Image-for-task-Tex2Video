# Save frames to dir

Becase we take from old GIF dataset, it have short, static and low resolution movis.

We have to select more good movies!

```python
dir_name = '/content/images'
try:
  os.mkdir(dir_name)
except: pass
```

```python
import requests
import numpy as np
import PIL

# what separete
LEN_MOVI = 100
wait_movis = 500
MIN_SIZE = 300
path_gif = '/content/temp.gif' # save gif path
take_min = 15 # q_ty take frame from gif
take_max = LEN_MOVI+1

# for save information to dataframe
select_movis = 0
pil_df = []

for id_matchs in range(len(matchs)):
  bad_movi = 0
  low_movi = 0
  error = 0
  id_movi = matchs[id_matchs][0]
  caption = train_df.loc[id_movi][1]
  url =  train_df.loc[id_movi][0]

  try:
    #download gif
    with open(path_gif, 'wb') as f:
      f.write(requests.get(url).content)

    # open gif
    gif_img = PIL.Image.open(path_gif)

    # take original frames
    frames = [frame.convert('RGB') for frame in PIL.ImageSequence.Iterator(gif_img)]
    width, height = frames[0].size
  except:
    print(f'Non or bad gif')
    error = 1
    frame = []

  if len(frames) < take_min:
    if error == 1: error = 0
    else: print(f'Small gif {id_movi}')

  # select movi not shot than take_min
  else:
    # select movi not low than MIN_SIZE
    if width < MIN_SIZE  or  height < MIN_SIZE:
       print(f'Movi {id_movi} have low H or W')
       low_movi = 1

    if len(frames) > take_max and not low_movi:
      # chack some statick movi
      dif_frames = [(frame - np.array(frames[0])) for frame in frames[1:]]
      dif_frames = np.array(dif_frames)

      for k in range(dif_frames.shape[0]):
          if dif_frames[k].sum() == 0:
            bad_movi+=1
      if not bad_movi:
        frames = frames[:take_max]
        print(f'Cut movi {id_movi}')
    else:
      if bad_movi:
        print(f'Movi {id_movi} have static frames')


    if not bad_movi and not low_movi:
        # save frames and make
        for i in range(len(frames)):
            id_frame = i
            img_path = os.path.join(dir_name, f'{id_movi}_{id_frame}.jpg')
            frames[i].save(img_path)
            pil_df.append([id_movi, f"{width},{height}", id_frame, [width, height], caption, img_path, patterns_to_search])

        select_movis += 1
        print(f'Movi {id_movi} have len = {len(frames)} take all move with W_{width} x H_{height} selected_{select_movis}')

    if select_movis > wait_movis: break

pil_df = pd.DataFrame(pil_df)
pil_df.columns = ['id_movi', "w x h", 'id_frame', 'size_frame', 'caption_movi', 'paths','patterns' ]
pil_df.to_csv('frames_info.csv', index=False)
```

```python
import glob
import PIL
import base64
from IPython import display

def make_gif(set_dir_pil,
             out_name = "my_awesome.gif",
             duration = 10,
             loop=0,
             ext = '/*.JPG'):

    if type(set_dir_pil) == str:
       frames = [PIL.Image.open(image) for image in glob.glob(f"{set_dir_pil}{ext}")]

    elif type(set_dir_pil) == list: frames = set_dir_pil

    frame_one = frames[0]
    frame_one.save(out_name, format="GIF", append_images=frames,
               save_all=True, optimize=False, duration=duration, loop=loop)


def show_gif(fname):
    with open(fname, 'rb') as fd:
        b64 = base64.b64encode(fd.read()).decode('ascii')
    return display.HTML(f'<img src="data:image/gif;base64,{b64}" />')
```

## flush_memory

```python
import torch
def flush_memory():
    import gc
    gc.collect()
    torch.cuda.empty_cache()
    torch.cuda.ipc_collect()
    with torch.no_grad():
        for _ in range(3):
          torch.cuda.empty_cache()
          torch.cuda.ipc_collect()


def replace_in_memory(device, to_gpu = [], to_cpu = [], ):
  if len(to_cpu):
    for obj in to_cpu:
      if obj.device != 'cpu': obj.to('cpu')

  flush_memory()

  if len(to_gpu):
    for obj in to_gpu:
      if obj.device != 'gpu': obj.to(device)
  flush_memory()
```

## get_embeds

```python
def get_image_embeds(pil_img, to_square = None, recize = []):

    if to_square:
      width, height = pil_img.size   # Get dimensions
      square_size = min(width, height)

      left = (width - square_size)/2
      top = (height - square_size)/2
      right = (width + square_size)/2
      bottom = (height + square_size)/2

      # Crop the center of the image
      pil_img.crop((left, top, right, bottom))

    if len(recize):
      pil_img = pil_img.resize(recize, PIL.Image.BICUBIC)

    with torch.no_grad():
      inputs = prior_T2I.image_processor(pil_img,
                                        return_tensors="pt")
      outputs = prior_T2I.image_encoder(**inputs.to(DEVICE)).image_embeds
      del inputs
      flush_memory()
    return outputs

def get_text_embeds(text, num_steps_inf = 30, seed = 0):
    return prior_T2I(prompt=text,
                      num_inference_steps=num_steps_inf,
                      num_images_per_prompt=1,
                      generator =  torch.manual_seed(seed)
                      ).image_embeds
```

```python
text = "George came home late last night"

prior_tokenizer = prior_T2I.tokenizer

inputs = prior_tokenizer([text],
                          padding="max_length",
                          max_length=prior_tokenizer.model_max_length,
                          truncation=True,
                          return_tensors="pt"
                          )
outputs = prior_T2I.text_encoder(**inputs.to(DEVICE))
```

# Collect dataset

prepare dataset with embedings images and texts

```python
# отрисовка прохождения цикла
from tqdm.notebook import tqdm

HEIGHT = 768
WIDTH = 768
pil_ids_movi_unique = pil_df.id_movi.unique()#[:10]
all_data = []

with torch.no_grad(): # prepare data for model and losses with no_grad
    for id_movi in tqdm(pil_ids_movi_unique, unit =" movi "):
        frames_pathes = pil_df[pil_df.id_movi == id_movi].paths
        frames_ids = pil_df[pil_df.id_movi == id_movi].id_frame.values
        # for normal way take before last
        pathes = frames_pathes[:-1]

        image_embeds = []
        ids_frame = []
        for i, path in enumerate(pathes):
            # seperate all start and random other frames
            if not i or ( i and np.random.random()<1) or i == len(pathes)-1:
                pil_image = PIL.Image.open(path)
                image_embeds.append(get_image_embeds(pil_image,
                                            to_square=True,
                                            recize=(HEIGHT, WIDTH)).cpu())
                # take label frames
                ids_frame.append(int(frames_ids[i]))

        if len(ids_frame):
            caption = pil_df[pil_df.id_movi == id_movi].caption_movi.values[0]

            inputs = prior_tokenizer([caption],
                                  padding="max_length",
                                  max_length=prior_tokenizer.model_max_length,
                                  truncation=True,
                                  return_tensors="pt"
                                  ).to(DEVICE)

            dict_movi = dict()
            dict_movi['id_movi'] = id_movi
            dict_movi['img_embeds'] = image_embeds
            dict_movi['last_hidden_state'] = prior_T2I.text_encoder(**inputs).last_hidden_state.cpu()
            dict_movi['text_embed'] = get_text_embeds(caption).cpu()
            dict_movi['ids_frames'] = ids_frame
            del(image_embeds, inputs)
            flush_memory()
            all_data.append(dict_movi)
```
